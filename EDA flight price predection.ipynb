# flight price predection(EDA and Feature engineering)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel("Data_Train.xlsx")
train_df.head()
```

```python
test_df=pd.read_excel("Test_set.xlsx")
```

```python
test_df.head()
```

```python
final_df=pd.concat([train_df,test_df])
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
#feature engineering process
final_df["Date"]=final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Month"]=final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Year"]=final_df["Date_of_Journey"].str.split("/").str[2]
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
#converting data type of mmonth date day feature to int
final_df[["Date","Month","Year"]]=final_df[["Date","Month","Year"]].astype(int)
```

```python
final_df.dtypes
```

```python
final_df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.columns
```

```python
final_df.head(10)
```

```python
final_df["Arrival_Time"]=final_df["Arrival_Time"].str.split(" ").str[0]
```

```python
final_df.head(10)
```

```python
final_df["Arrival_hour"]=final_df["Arrival_Time"].str.split(":").str[0]
final_df["Arrival_min"]=final_df["Arrival_Time"].str.split(":").str[1]
```

```python
final_df.head()
```

```python
final_df["Arrival_hour"]=final_df["Arrival_hour"].astype(int)
final_df["Arrival_min"]=final_df["Arrival_min"].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
 final_df["Dep_Hrs"]=final_df["Dep_Time"].str.split(":").str[0]
final_df["Dep_min"]=final_df["Dep_Time"].str.split(":").str[1]   
```

```python
final_df["Dep_Hrs"]=final_df["Dep_Hrs"].astype(int)
final_df["Dep_min"]=final_df["Dep_min"].astype(int)
```

```python
final_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df["Total_Stops"].unique()
```

```python
final_df["Total_Stops"]=final_df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4})
```

```python
final_df.head()
```

```python
final_df[final_df["Total_Stops"].isnull()]
```

```python
final_df.drop("Route",axis=1,inplace=True)
```

```python
final_df["Additional_Info"].unique()
```

```python
#converting duration into  minutes
final_df["Duration_Hr"]=final_df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
final_df[final_df["Duration_Hr"]=="5m"]
```

```python
final_df.drop([6474,2660],axis=0,inplace=True)
```

```python
final_df["Duration_Hr"]=final_df["Duration_Hr"].astype(int)
```

```python
final_df.drop("Duration",axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head(1)
```

```python
final_df["Airline"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencode=LabelEncoder()
```

```python
final_df["Airline"]=labelencode.fit_transform(final_df["Airline"])
final_df["Source"]=labelencode.fit_transform(final_df["Source"])
final_df["Destination"]=labelencode.fit_transform(final_df["Destination"])
final_df["Additional_Info"]=labelencode.fit_transform(final_df["Additional_Info"])
```
